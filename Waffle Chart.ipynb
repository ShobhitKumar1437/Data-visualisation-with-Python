```python
   #  WAFFLE CHART

import matplotlib.pyplot as plt 
import pandas as pd 
from pywaffle import Waffle
data = {'Phone':['Mi','Apple','Samsung','Nokia','Realme'],
        'Stock':[44,12,8,5,3]}
df=pd.DataFrame(data)
fig=plt.figure(figsize=(8,9),FigureClass=Waffle,rows=5,values=df.Stock,labels=list(df.Phone))
plt.title('Waffle Chart')
plt.show()
```
