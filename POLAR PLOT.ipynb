```python
# POLAR PLOT

import matplotlib.pyplot as plt
import numpy as np

wd = [30,60,90,120,150,180,210,240,270,300,330]
ws = [5,7,8,6,4,3,5,7,9,6,4]

theta = np.radians(wd)


plt.figure(figsize=(10,9))
plt.polar(theta,ws,linestyle='-',color ='green',marker = '*',markersize=30,markerfacecolor='red')
plt.title("Polar Plot")
plt.xlabel("Wind Direction")
plt.ylabel("Wind( Speed")
plt.tight_layout()
plt.show()
```
