```python
import matplotlib.pyplot as plt
tata = [5,6,7,9,8,-2,4]
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
plt.plot(days,tata)
plt.title("Stoke prices as per days in week")
plt.xlabel("Days")
plt.ylabel("Stoke Price")
plt.show()
```

```python
import matplotlib.pyplot as plt
months = ["Jan","fab","mar","apr","may","jun","jul"]
marks = [55,78,68,76,81,45,52]
plt.plot(months,marks)
plt.title("Student marks as per months")
plt.xlabel("Months")
plt.ylabel("Marks")
plt.show()
```
