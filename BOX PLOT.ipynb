```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

data=[78,82,85,88,90,92,93,95,97,100,65,72,150,40]
df = pd.DataFrame(data , columns= ['age'])

ile25 = df.age.quantile(0.25)
ile75 = df.age.quantile(0.75)


iqr = (ile75-ile25)

ul = ile75+(iqr*1.5)
ll = ile25-(iqr*1.5)


df = df[df.age<=ul]
df = df[df.age>=ll]

plt.boxplot(df)
plt.show()





```
