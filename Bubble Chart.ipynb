```python
#  BUBBLE CHART

import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5000,6000,7000,6000,9000]
z=[15,30,82,68,95]
bs=[]
for i in z:
    a=i*10
    bs.append(a)
    print(bs)
plt.figure(figsize=(8,6))
plt.scatter(x,y,s=bs,c='orange',alpha=0.8,edgecolors='black')
plt.xlabel("Polution")
plt.ylabel("Income")
plt.title("Bubble Plot")
plt.show()
```
