```python
#  contour f 

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
img = Image.open(r"E:\Photo\IMG_20231201_013139_409.jpg").convert('L')
image_data = np.array(img)
plt.figure(figsize=(10,9))
level = np.linspace(1,10,10)
c = plt.contourf(image_data,levels =level,cmap = 'Reds')
plt.imshow(image_data,cmap='gray',alpha=0.5)
plt.colorbar(c,label='intensity')
plt.tight_layout()
plt.show()
```

```python
# Only contour
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
img = Image.open(r"E:\Photo\IMG_20231201_013139_409.jpg").convert('L')
image_data = np.array(img)
plt.figure(figsize=(10,9))
level = np.linspace(1,10,10)
c = plt.contour(image_data,levels =level,cmap = 'Reds')
plt.imshow(image_data,cmap='YlGn',alpha=0.5)
plt.colorbar(c,label='intensity')
plt.tight_layout()
plt.show()
```
